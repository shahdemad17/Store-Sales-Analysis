**Build Data Model ,Data Cleaning and Preprocessing..**

---the first step in the process is importing pandas library and data file to start cleaning data

```python
import pandas as pd
df= pd.read_csv('Sales Data.csv')
df= pd.DataFrame(df)
df.head(5)
df.info()
df.dtypes

```

**--The Second Step is Data Cleaning Process**
by Removing Duplicates and Filling Missing Values
,For The Missing Values There is only 11 Nulls so We Can Remove Them Easly Without Affecting Analysis Process..

```python
df.drop_duplicates(inplace=True)
df.info
df.isnull().sum()
df.dropna(inplace=True)
df.info()
```

**Converting Float values of Sales to Integer Values To make it easier in calculations .**

```python
df['Sales']=df['Sales'].round().astype(int)
df.head(3)

```

**convert shipdate and order date from string to date and extract shipment days.**

```python
print(type("Ship Date"))
df["Ship Date"] = pd.to_datetime(df["Ship Date"], format="%d/%m/%Y")
df["Order Date"] = pd.to_datetime(df["Order Date"], format="%d/%m/%Y")
df['Shipment Days']=df['Ship Date']-df['Order Date']
print(type('Shipment Days'))
# Extract only the numeric part directly without the word 'days'
#df['Shipment Days'] = df['Shipment Days'].str.extract('(\d+)')
#df['Shipment Days'] = pd.to_numeric(df['Shipment Days'])
# Extract only the numeric part and convert it to an integer
df['Shipment Days'] = df['Shipment Days'].dt.days.astype(int)
#df["Year"] = df["Ship Date"].dt.year
#df["Month"] =df["Ship Date"].dt.month
#df.drop('Day',axis=1,inplace=True)
#df.drop('Day_Name',axis=1,inplace=True)
df.head(3)
```
